## About the Data

The data has 4 independent features:
* Sepal Length in Cm
* Sepal Width in Cm
* Petal Length in Cm	
* Petal Width in Cm

And 1 dependent feature:
* Species: 
having the following unique labels:
1. Iris-setosa
2. Iris-versicolor
3. Iris-virginica


## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

The first column is Id which need not be included in X array

```
dataset = pd.read_csv('Dataset_Iris.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```
dataset
```

```
X
```

```
y
```

## Encoding Dependent y column

The y has all its values as strings (labels) which have to be encoded and converted to integers using Label Encoder as follows

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

```
le.inverse_transform([0])
```

## Splitting the dataset into the Training set and Test set

75% of the dataset is used for training the model and 25% for testing it

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
print(X_train)
```

```
print(y_train)
```

```
print(X_test)
```

```
print(y_test)
```

## Training the Decision Tree Classification model on the Training set

```
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

The accuracy score is 97.37%. Only 1 result has been wrongly predicted.
